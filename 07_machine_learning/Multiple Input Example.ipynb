```python
import numpy as np
from tensorflow.keras.models import Sequential #  Sequential() 
from tensorflow.keras.layers import Dense #  Dense() 
from tensorflow.keras import optimizers #   

#    3. , input_dim 3.
X = np.array([[70,85,11],[71,89,18],[50,80,20],[99,20,10],[50,10,10]]) # , , 

#    1. , output_dim 1.
y = np.array([73,82,72,57,34]) #  

model=Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))

# (learning rate, lr) 0.00001 .
sgd=optimizers.SGD(lr=0.00001)

#  (Loss function)  mse .
model.compile(optimizer = sgd ,loss='mse',metrics=['mse'])

#  X y     2,000 .
model.fit(X,y, batch_size=1, epochs=2000, shuffle=False)
```

```python
print(model.predict(X))
```

```python
X_test = np.array([[20,99,10],[40,50,20]]) #  58 56  .
print(model.predict(X_test))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential #  Sequential() 
from tensorflow.keras.layers import Dense #  Dense() 
from tensorflow.keras import optimizers #   

#    2. , input_dim 2.
X=np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

#    1. , output_dim 1.
y=np.array([0, 1, 1, 1])

model=Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid')) #    2.

# sgd   .
#  (Loss function) binary_crossentropy(  ) .
model.compile(optimizer='sgd' ,loss='binary_crossentropy',metrics=['binary_accuracy'])

#  X y     800 .
model.fit(X, y, batch_size=1, epochs=800, shuffle=False)
```

```python
print(model.predict(X))
```
